# LightFM Hybrid Recommend
    Thư viện LightFM sử dụng mô hình nhúng ma trận (matrix factorization) --> Hỗ trợ 3 phương pháp: Collaborative, Content, và Hybrid
## Ưu điểm của Hybrid Recommender trong LightFM
    - Giải quyết vấn đề cold-start: Khi có thêm thông tin về sản phẩm và người dùng, hệ thống có thể đề xuất ngay cả khi chưa có nhiều tương tác
    - Tận dụng cả thông tin cộng tác và nội dung: Cải thiện độ chính xác bằng cách kết hợp dữ liệu tương tác với thông tin đặc trưng
    - Tối ưu hoá hiệu xuất: LightFM sử dụng thuật toán học sâu (SGD, Adagrad) để huấn luyện mô hình nhanh hơn so với các phương pháp truyền thống
