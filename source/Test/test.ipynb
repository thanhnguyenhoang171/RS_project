```python
import numpy as np
from lightfm import LightFM
from lightfm.data import Dataset
```

```python
from scipy.sparse import coo_matrix

# Dữ liệu giả lập: người dùng, sản phẩm, và rating
users = np.array([0, 0, 1, 1, 2, 2])
items = np.array([0, 1, 1, 2, 2, 3])
ratings = np.array([5, 4, 4, 5, 3, 5])

# Chuyển đổi sang ma trận sparse
num_users = len(set(users))
num_items = len(set(items))
interaction_matrix = coo_matrix((ratings, (users, items)), shape=(num_users, num_items))
```

```python
# Khởi tạo model với loss function (Warp, BPR, Logistic)
model = LightFM(loss="warp")

# Huấn luyện mô hình
model.fit(interaction_matrix, epochs=30, num_threads=4)
```

```python
def recommend(model, user_id, num_items=3):
    scores = model.predict(user_id, np.arange(num_items))
    top_items = np.argsort(-scores)
    return top_items


print(recommend(model, user_id=1, num_items=4))
```
